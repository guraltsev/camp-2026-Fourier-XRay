# Interaction of Fourier modes

Fourier analysis works because the waves
\begin{equation}
\sin(2\pi n x), \qquad \cos(2\pi m x), \qquad n,m∈\mathbb{N}
\end{equation}

interact in an organized way.

We have already seen that it is hard to understand interactions between Fourier modes under addition. Let us study how Fourier modes behave under *multiplication*.

## Sine against sine

We start by comparing two sines at two different frequencies. Watch the product curve as the parameters change.

````python
fig1 = figure("Sine products", new=True)
fig1.view.home_range = {"x": (-1 / 2, 1 / 2), "y": (-1.2, 1.2)}
fig1.show()

fig1.parameters({b[1]: {"value": 0, "min": 0, "max": 20, "step": 1},
                 b[2]: {"value": 0, "min": 0, "max": 20, "step": 1}})

fig1.plot( sin(2 * pi * b[1] * x), x, name="sine1",  style={"width": 3, "color":"blue", "opacity": 0.2}, samples=1001)
fig1.plot(sin(2 * pi * b[2] * x), x, name="sine2",  style={"width": 3, "color":"green",  "opacity": 0.2}, samples=1001)
fig1.plot(sin(2 * pi * b[1] * x)*sin(2 * pi * b[2] * x), x, name="product", style={"width": 2,"color":"red"}, samples=1001)
````

## Cosing against Cosine

Next, we compare two cosines at two different frequencies. Watch the product curve as the parameters change.

````python
fig2 = figure("Cos products", new=True)
fig2.view.home_range = {"x": (-1 / 2, 1 / 2), "y": (-1.2, 1.2)}
fig2.show()

fig2.parameters({a[1]: {"value": 0, "min": 0, "max": 20, "step": 1},
                 a[2]: {"value": 0, "min": 0, "max": 20, "step": 1}})

fig2.plot( cos(2 * pi * a[1] * x), x, name="cos1",  style={"width": 3, "color":"blue", "opacity": 0.2}, samples=1001)
fig2.plot(cos(2 * pi * a[2] * x), x, name="cos2",  style={"width": 3, "color":"green",  "opacity": 0.2}, samples=1001)
fig2.plot(cos(2 * pi * a[1] * x)*cos(2 * pi * a[2] * x), x, name="product", style={"width": 2,"color":"red"}, samples=1001)
````

## Sine against cosine

Finally, compare sine modes with cosine modes.

````python
fig3 = figure("Cos products", new=True)
fig3.view.home_range = {"x": (-1 / 2, 1 / 2), "y": (-1.2, 1.2)}
fig3.show()

fig3.parameters({a: {"value": 0, "min": 0, "max": 20, "step": 1},
                 b: {"value": 0, "min": 0, "max": 20, "step": 1}})

fig3.plot( cos(2 * pi * a * x), x, name="cos1",  style={"width": 3, "color":"blue", "opacity": 0.2}, samples=1001)
fig3.plot(sin(2 * pi * b * x), x, name="sin2",  style={"width": 3, "color":"green",  "opacity": 0.2}, samples=1001)
fig3.plot(cos(2 * pi * a * x)*sin(2 * pi * b * x), x, name="product", style={"width": 2,"color":"red"}, samples=1001)
````
