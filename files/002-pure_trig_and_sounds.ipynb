# Periodic functions and how they sound



##  Pure trigonometric waves

A sine wave is one of the simplest periodic signals we can draw and hear.

\begin{equation*}
b \sin(2\pi k x).
\end{equation*}

- $b$ is called the **amplitude** - it controls the vertical size of the wave. 
- $k$ is the **frequency** - it controls the rate of oscillation.

When $x$ is time in seconds, $k$ is measured in Hertz.

Move the sliders to study the effect of the parameters on the sound.

````python
fig1 = figure("Trig waves", new=True)
fig1.view.range = {"x": (0, 0.01), "y": (-1.1, 1.1)}
fig1.show()

fig1.plot(
    b * sin(2 * pi * k * x),
    x,
    name="sin-wave",
    style={"width": 3},
    samples=1200,
)
fig1.parameters({
    b:{"value":0.7,"min": -1, "max":1, "step":0.01},
    k:{"value":220, "min": 50, "max":220*2**5, "step":0.1},
                })
````

Amplitude is a scale factor: it stretches the same shape up and down. Frequency is a repetition count: it changes how often the same shape repeats. A pure sine wave has only one frequency.

## Other sound waves

A sine wave has a smooth round shape. Other periodic waves repeat just as regularly, but their corners or jumps change the sound.

The next figure compares three waves with the same amplitude and fundamental frequency. Use the legend sound buttons to listen to one curve at a time.

````python
# Periodic square and triangle waves with unit amplitude.
Sq = Function("Sq")(x) @EqDef@ (
    If(x-floor(x) > 1/2).Then(1)
    .If(x-floor(x) < 1/2).Then(-1)
    .Otherwise(0)
)
Tr = Function("Tr")(x) @EqDef@ (
    If(x-floor(x)>1/2).Then(1+4*(x-floor(x)-1))
    .If(x-floor(x) < 1/2).Then(1-4*(x-floor(x)))
    .Otherwise(1)
)

md(r"""
The square wave and triangle wave used below are

$$
\mathrm{Sq}(x) := {{ Sq(x)>>Expand }},
\qquad
\mathrm{Tr}(x) := {{ Tr(x)>>Expand }}.
$$
""")
````

````python
fig2 = figure("Compare wave shapes", new=True)
fig2.view.home_range = {"x": (0, 0.01), "y": (-1.2, 1.2)}
fig2.show()

fig2.plot(
    b * sin(2 * pi * k * x),
    x,
    name="sine",
    label=r"$b\sin(2 \pi k x)$",
    style={"width": 3},
    samples=1600,
)
fig2.plot(
    b * Sq(k * x),
    x,
    name="square",
    label=r"$b \mathrm{Sq}(k x)$",
    style={"width": 2,},
    samples=1600,
)
fig2.plot(
    b * Tr(k * x),
    x,
    name="triangle",
    label=r"$b\mathrm{Tr}(k x)$",
    style={"width": 2,},
    samples=1600,
)

fig2.parameters({
    b: {"value": 0.7, "min": -1, "max": 1, "step": 0.01},
    k: {"value": 220, "min": 50, "max": 220 * 2**5, "step": 0.1},
})
````

## Building sounds by adding pure waves

A more flexible sound comes from adding many pure waves. The next model uses twenty harmonics of one fundamental frequency:

$$
M(x)=\sum_{n=1}^{20} c_n\sin(2\pi n f_0 x).
$$

Each slider controls one harmonic amplitude. Start with one or two sliders, then add more and listen for how the tone changes.

````python
# Twenty-harmonic sine model for experimenting with timbre.
fig3 = figure("Twenty-wave sound mixer", new=True)
fig3.view.home_range = {"x": (0, 0.01), "y": (-2.2, 2.2)}
fig3.show()


Nmax = 20
M = Function("M")(x) @EqDef@ (Sum(
    b[n]*sin(2*pi*k*n*x)
    ,(n,1,Nmax)
))


fig3.plot(
    M(x),
    x,
    name="mix",
    style={"width": 3},
    samples=2400,
)
fig3.sound.normalization = True

fig3.parameters({
    k: {"value": 220, "min": 50, "max": 220 * 2**5, "step": 0.1},
})
````
