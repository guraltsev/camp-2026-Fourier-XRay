# Polynomial interpolation and unstable coefficients
For fixed $N$, choose the $N+1$ equally spaced points on $[0,1]$:
\begin{equation*}
X_i := \frac{i}{N}
\end{equation*}

There is a unique polynomial of degree at most $N$ through any prescribed values at these $N+1$ points. Let 
\begin{equation}\label{eq:polynomial}
P(x)=\sum_{j=0}^{N} A_j x^j
\end{equation}

The interpolation problem is:

> Given the values $Y_i$ at the points $X_i$ compute the coefficients of the polynomial having the given values at the given points.

In practice, find $A_0,\ldots, A_N$ such that $P(X_i)=Y_i$ for every $i∈\{1,\ldots,N\}$. 



## Finding the coefficients

Using the notation {eq}`eq:polynomial`, the condition $Y_i=P(X_i)$ becomes 
\begin{equation*} 
Y_i=\sum_{j=0}^N X_i^j A_j
\end{equation*}

We rewrite this as a matrix equation. Denote $V_{ij}=X_i^j$ so that the condition $Y_i=P(X_i)$ becomes 
\begin{equation*}
Y_i=\sum_{j=0}^{N} V_{ij} A_j.
\end{equation*}


We immediately recogize a matrix by vector multiplication. Our condition is 
\begin{equation}\label{eq:interpolation-matrix-equation}
Y= V A.
\end{equation}
Therefore, to find the coefficients $A=\begin{pmatrix}a_0\\\vdots \\ a_N\end{pmatrix}$
we need to compute $ A=V^{-1}Y $.


## An example with a small number of points

Let us set the maximum number of points $N_{max}$ first to $5$. 

At $N_{max}=5$, the points are $\{0,1/4,1/2,3/4,1\}$. 

````python
# Default value for this notebook
Nmax = 5
````

First we compute the matrix $V$ with $N=N_{max}$.

````python
# Compute V
V = Matrix(Nmax+1, Nmax+1, lambda i,j:(i/Nmax)**j)
r""" $$V= {{ V }} $$ """ >> md
````

Therefore we get 

````python
# Display V^{-1}
r"""
$$
V^{-1} = {{ V**(-1) }}.
$$
""" >> md
````

The coefficients of the polynomial $P(x)$ therefore are

````python
A_of_Y = (V ** (-1)) * Matrix(Nmax+1, 1, lambda i,j: Y[i])

r"""
$$
A
=\begin{pmatrix}A_0\\\vdots \\ A_{N_{max}}\end{pmatrix} 
= {{ A_of_Y}}
$$
""">> md
````

Therefore, our polynomial is 

````python
P = Function("P")(x) @EqDef@ (Sum(A_of_Y[j,0]*x**j, (j,0,Nmax)) >> DoIt)
P(x)>> md
````

### Interpolating - a live experiment

Let us keep the equally spaced points $X_i$ fixed and see how the interpolating polynomial depends on the data values $Y_0,\ldots,Y_{N_{max}}$. We will plot $P(x)$ and compare it with the polynomial $Q(x)$ given by 


````python
Q = Function("Q")(x) @ EqDef @ (2 * x**4 + x**3 - x**2 - x + 1)
r"""
$$Q(x)={{ Q(x) }}$$
""" >> md
````

````python
fig1 = figure("Polynomial interpolation", new=True)
fig1.view.home_range = {"x": (0, 1), "y": (-3.0, 4.0)}
fig1.show()

fig1.plot(
        P(x), x,
        name="P",
        label=r"$P(x)$",
        style={"width": 3},
    )
fig1.plot(
        Q(x), x,
        name="Q",
        label=r"$Q(x)$",
        style={"width": 3, "opacity":0.4},
    )
fig1.list_plot(
        (k/Nmax, Y[k]),
        (k, 0, Nmax + 1),
        name="data",
        label=r"$(x_i,y_i)$",
    )
````
