# Euler's formula from tiny turns

Euler's formula says

$$
e^{it}=\cos(t)+i\sin(t).
$$

This notebook looks at the formula through the finite product definition of the exponential. Instead of jumping straight to the limit, we draw the partial products.

## A finite product in the complex plane

For a large integer $N$, the exponential is approximated by

$$
e^{it}\approx \left(1+i\frac{t}{N}\right)^N.
$$

The intermediate powers

$$
\left(1+i\frac{t}{N}\right)^n,\qquad n=0,1,\ldots,N,
$$

show the whole walk, not just the final point.

```python
# Give the finite product a symbolic name so the plot below has a formula.
Z = Function("Z")(n, t, N) @EqDef@ ((1 + I * t / N) ** n)
X = Function("X")(n, t, N) @EqDef@ re(Z(n, t, N))
Y = Function("Y")(n, t, N) @EqDef@ im(Z(n, t, N))

r"""
We list-plot the points
$$
(X(n,t,N),Y(n,t,N)) = ({{ X(n,t,N) }}, {{ Y(n,t,N) }}),
\qquad 0\le n\le N.
$$

The red point is
$$
(\cos(t),\sin(t)).
$$
""" >> md
```

## List-plot the real and imaginary parts

Move $N$ to change the number of small complex multiplications. Move $t$ to change the angle we are aiming for.

```python
import ipywidgets as widgets
from IPython.display import display


def finite_euler_points(t_value, N_value):
    """Return the finite product points as real-imaginary coordinates."""

    step = 1 + 1j * t_value / N_value
    z_value = 1 + 0j
    points = []

    for _ in range(N_value + 1):
        points.append((z_value.real, z_value.imag))
        z_value *= step

    return points


def update_euler_figure(change=None):
    t_value = float(t_slider.value)
    N_value = int(N_slider.value)
    points = finite_euler_points(t_value, N_value)
    endpoint = points[-1]
    target = (float(cos(t_value)), float(sin(t_value)))
    distance = float(sqrt((endpoint[0] - target[0])**2 + (endpoint[1] - target[1])**2))

    euler_fig.list_plot(
        points,
        name="finite powers",
        label=r"$\left(1+i\frac{t}{N}\right)^n$",
        style={"color": "#2563eb", "width": 2},
    )
    euler_fig.list_plot(
        [target],
        name="Euler target",
        label=r"$(\cos(t),\sin(t))$",
        style={"color": "#dc2626", "width": 5},
    )
    euler_fig.list_plot(
        [(1, 0)],
        name="start",
        label=r"$1+0i$",
        style={"color": "#16a34a", "width": 5},
    )
    readout.value = (
        f"<b>N</b> = {N_value} &nbsp; "
        f"<b>t</b> = {t_value:.3f} &nbsp; "
        f"endpoint = ({endpoint[0]:.5f}, {endpoint[1]:.5f}) &nbsp; "
        f"distance to target = {distance:.2e}"
    )


euler_fig = figure("Euler finite product", new=True)
euler_fig.view.home_range = {"x": (-1.75, 1.75), "y": (-1.75, 1.75)}
euler_fig.show()

N_slider = widgets.IntSlider(
    value=32,
    min=4,
    max=256,
    step=1,
    description="N",
    continuous_update=False,
)
t_slider = widgets.FloatSlider(
    value=float(pi / 2),
    min=float(-2 * pi),
    max=float(2 * pi),
    step=float(pi / 64),
    readout_format=".3f",
    description="t",
    continuous_update=False,
)
readout = widgets.HTML()

N_slider.observe(update_euler_figure, names="value")
t_slider.observe(update_euler_figure, names="value")
update_euler_figure()

display(widgets.VBox([widgets.HBox([N_slider, t_slider]), readout]))
```

```python
# The size and angle of the final product explain the limiting point.
Length = Function("Length")(t, N) @EqDef@ ((1 + t**2 / N**2) ** (N / 2))
Angle = Function("Angle")(t, N) @EqDef@ (N * atan(t / N))

r"""
The last blue point has length
$$
{{ Length(t,N) }}
$$
and angle
$$
{{ Angle(t,N) }}.
$$

As $N$ grows, these approach $1$ and $t$.
""" >> md
```

## Why the picture becomes Euler's formula

Multiplication by $1+i\frac{t}{N}$ is almost a tiny rotation by angle $\frac{t}{N}$. After $N$ steps, the accumulated angle is almost $t$.

Its length is

$$
\left|1+i\frac{t}{N}\right|^N=\left(1+\frac{t^2}{N^2}\right)^{N/2},
$$

which approaches $1$ as $N$ grows. The limiting endpoint therefore has angle $t$ and length $1$, so its coordinates are $(\cos(t),\sin(t))$.
