# Equations and LaTeX

Use structural notation when the main payoff is a readable equation on the page and a clean LaTeX form underneath.


## Why this notation helps

When indices and function labels are part of the mathematics, notebook display and LaTeX output should preserve that structure instead of forcing you to decode it from names. This is especially noticeable in discrete schemes and residuals.


## A finite-difference style relation

Start with an equation that uses both a time index and a component index.


````python
u, h = symbols("u h")
scheme = ( u[t + 1, i] @Eq@ (u[t, i] + h * (u[t, i + 1] - u[t, i])) )
scheme >> md
pass
````

## The same relation as rendered LaTeX

Structural indices flow straight through to the rendered form.


````python
"""
The same equation renders as notebook math and as reusable LaTeX source.
$$
{{ scheme }}
$$
*LaTeX:*
```text
{{ latex(scheme) }}
```
""" >> md
pass
````

## Multi-letter names stay readable too

The atom-level LaTeX patch makes plain names such as `speed` render as symbols rather than raw identifier text.


````python
speed = Symbol("speed")
speed_balance = speed[t + 1] @Eq@ (speed[t] + alpha)
speed_balance >> md
pass
````

````python
speed_latex = latex(speed_balance)

"""
The named-speed equation keeps the readable atom in both rendered math and LaTeX source.
$$
{{ speed_balance }}
$$
*LaTeX*:
```text
{{ speed_latex }}
```
""" >> md
pass
````

## Recap

Structural notation improves both the mathematics you manipulate and the mathematics you show. The same object stays readable in notebook output, in equations, and in generated LaTeX.
