# Composing structured expressions

Build mathematical expressions from symbols, indexed quantities, and function families without losing readability.


## Start from a structured update

A single equation can already combine several topic areas.


```python
x[t + 1] @Eq@ (F[t](x[t]) + b[t])
```

## Write the residual you would simplify or solve

Residual form is often the object you manipulate next.


```python
x[t + 1] - F[t](x[t]) - b[t]
```

## Substitute one label at a time

Because the indices are structural, relabeling is straightforward.


```python
(x[t + 1] - F[t](x[t]) - b[t]) >> Subs({t:n})
```

## Simplify without peeling apart the notation by hand

Ordinary expression operations still work on the structured pieces.


```python
(x[t] - y[t] + y[t]) >> Simplify
```

## Nest indexed maps when the model asks for it

Composition stays legible even when the time label appears at several levels.


```python
F[t + 1](x) @Eq@ G[t](F[t](x))
```

## Recap

Composed expressions are where the notation earns its keep. You can build updates, residuals, substitutions, and nested maps without giving up either readability or symbolic manipulability.
