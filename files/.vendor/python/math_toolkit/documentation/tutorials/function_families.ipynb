# Function families

Use indexed function heads when the map itself changes with a label such as time, regime, or component.


## Why this notation helps

`F[t](x)` and `F(x[t])` say different things. The first means “the map changes with `t`.” The second means “the argument changes with `t`.” Structural notation lets you keep that distinction explicit.


## Two different ideas

Compare the two most common patterns directly.


```python
F[t](x)
```

```python
F(x[t])
```

## Following the label through substitutions

The label sits in a different place in each expression, so substitutions change different pieces of the mathematics.


```python
F[t](x) >> Subs({t: s})
```

```python
F(x[t]) >> Subs({t: s})
```

## Applied families behave like expressions

Once the indexed head is called, you can differentiate or place the result inside larger formulas.


```python
F[t](x) >> Diff(x)
```

## A time-varying update map

Here the update rule itself depends on the time label.


```python
x[t + 1] @Eq@ F[t](x[t])
```

## More than one label

Repeated brackets extend the label list rather than creating a new kind of object.


```python
F[t][i](x)
```

## Recap

Indexed heads are for changing maps. Indexed arguments are for changing inputs. Keeping those roles separate pays off immediately in substitutions, differentiation, and recurrence-style models.
