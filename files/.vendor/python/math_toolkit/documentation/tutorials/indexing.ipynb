# Indexing notation

Represent time steps, components, and labeled quantities as real structure instead of text hidden inside names.


## Why this notation helps

In exploratory symbolic work, labels like `t + 1` or `(i, j)` are part of the mathematics, not decoration. Structural indexing keeps those labels separate, readable, and available for later substitution or algebra.


## First contact

Start with the smallest meaningful indexed objects.


```python
x[t + 1]
```

```python
x[i, j]
```

## Building a discrete update

Once the indices are structural, recurrence-style notation reads like the mathematics you intended.


```python
x[t + 1] @Eq@ (A[t] * x[t] + b[t])
```

## Turning the update into a residual

Residual form is often the object you want for simplification, substitution, or later analysis.


```python
x[t + 1] - A[t] * x[t] - b[t]
```

## Indices stay symbolic

Because the label is a real symbolic part, you can replace it without rebuilding the whole expression by hand.


```python
(x[t + 1] - A[t] * x[t] - b[t]).subs(t, n)
```

## Indexed function heads

The same bracket notation works when the map itself varies with the label.


```python
x[t + 1] @Eq@ F[t](x[t])
```

## Recap

Structural indexing is most useful when the labels matter mathematically. It keeps recurrence relations, component notation, and time-varying maps readable while still leaving the labels available for later symbolic work.
