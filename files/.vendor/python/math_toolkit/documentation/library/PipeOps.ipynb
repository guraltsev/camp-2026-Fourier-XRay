# PipeOps

`expr >> Expand`  
`expr >> Simplify`  
`expr >> Subs({x: 2})`  
`expr >> Diff(x)`  
`expr >> DoIt`

`PipeOp` objects are immutable unary operators that can be piped with `>>`.
The `math_toolkit.pipeops.common_pipeops` module provides curated uppercase
symbolic operators such as `Expand`, `Simplify`, `Subs`, `Replace`, `Evalf`,
`Diff`, `DoIt`, `Series`, and `Taylor`.

Direct operators such as `Expand` receive the left-hand expression directly.
Payload-taking operators such as `Subs`, `Diff`, `Evalf`, and `Series`
capture one explicit payload first and then accept the expression from the
left side of the pipeline.
