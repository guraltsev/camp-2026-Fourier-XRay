# reset

Refresh the managed notebook names after you have reused `x`, `F`, or other convenience bindings in a working session.

`math_toolkit.notebook.reset()`  
`math_toolkit.notebook.reset(namespace=scope)`


```python
x = "temporary text"
math_toolkit.notebook.reset()
x[i + 1] @Eq@ (x[i] + 1)
```

## Core behavior

`reset()` writes the managed notebook bindings into a namespace without changing the package profile selected at import time. It is the quickest way to recover the default names after you have overwritten one of them during exploration.


## Common patterns

Use `reset()` when the notation disappeared because you reassigned a managed name in the current notebook.

## Options and Details

Pass an explicit namespace when helper code or tests need to refresh a mapping without touching notebook globals. Existing unrelated entries stay in place.

```python
scope = {"x": "sentinel", "custom": 123}
math_toolkit.notebook.reset(scope)
scope["x"][scope["i"] + 1] @Eq@ (scope["x"][scope["i"]] + 1)
```

```python
scope["custom"]
```

## Semantics

`reset()` restores the managed names, but it does not change which extension modules the package profile imported.


## Pitfalls

`reset()` is not a full workspace clear. It does not remove custom variables or undo unrelated side effects.
