# Function

Use function heads to represent symbolic maps, then apply them or index the head when the map itself varies.

`F`  
`F(x)`  
`F[t](x)`


```python
F(x + y) @Eq@ (F(x) + F(y))
```

## Core behavior

A function head names an operation without choosing its argument yet. That lets you write symbolic identities such as `F(x + y) = F(x) + F(y)` before you know what `F` will mean. Once you call the head, the result joins the broader expression system.


## Common patterns

Function heads are useful both for ordinary symbolic maps and for families of maps that carry structural labels.


```python
diff(F(x), x)
```

```python
x[t + 1] @Eq@ F[t](x[t])
```

## Semantics

`F` and `F[t]` are both callable heads. They name operations. Once you apply them, the result is an `Expression`. Use indexing on the head when the map varies with a label such as time, and use indexing on the argument when the input varies instead.


## Pitfalls

`F[t]` and `F(t)` mean different things. The first is an indexed function head; the second is an ordinary function applied to the argument `t`. Likewise, use structural indexing instead of encoding labels into the function name.
