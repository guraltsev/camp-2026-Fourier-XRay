# domain_plot

`domain_plot(...)` fills a two-dimensional region described by symbolic conditions or signed scalar expressions.

Preferred form:

```python
fig = figure("domain")
fig.show()
fig.domain_plot(condition, (x, xmin, xmax), (y, ymin, ymax))
```

````python
x, y, r = symbols("x y r")

fig = figure("domain-quick-payoff")
fig.show()
fig.parameters({r: 1.0})
fig.domain_plot(x**2 + y**2 < r**2, (x, -2, 2), (y, -2, 2), label="disk")
````

## Core behavior

`fig.domain_plot(condition, x_domain, y_domain)` plots where a symbolic Boolean condition is true. A signed scalar expression is also accepted: positive values describe the filled domain and zero describes the boundary.

The returned `PlotHandle` can update labels, styles, parameters, display through the parent figure, or remove the domain plot.

## Common patterns

**Boolean regions.** Use symbolic comparisons such as `x**2 + y**2 < 1` or `(x > 0) & (y > 0)`.

**Signed domains.** Use an expression such as `1 - x**2 - y**2` when the sign carries the geometry.

**Systems of conditions.** A finite list can combine Boolean conditions and signed expressions.

**Boundary control.** Use `boundary=False` when you want the filled region without its boundary trace.

````python
fig = figure("domain-system")
fig.show()
fig.domain_plot(
    [x > 0, y > 0, 1 - x - y],
    (x, -0.2, 1.2),
    (y, -0.2, 1.2),
    name="simplex",
    label="x >= 0, y >= 0, x + y <= 1",
)
````

## Options and Details

`samples=` accepts one integer for both axes or `(x_samples, y_samples)`.

`style=` is a nested dictionary with `domain` and `boundary` sections. The domain fill supports options such as `zsmooth`; the boundary supports options such as `smoothing`.

`boundary=` is a Boolean display switch for the boundary trace. It must be `True` or `False`.

`fig.parameters(...)` creates sliders for symbols that are not the x or y domain variables.

````python
fig = figure("domain-style")
fig.show()
fig.domain_plot(
    1 - x**2 - y**2,
    (x, -1.5, 1.5),
    (y, -1.5, 1.5),
    boundary=True,
    style={"domain": {"zsmooth": False}, "boundary": {"smoothing": 0}},
)
````

## Semantics

Boolean domains are true/false regions. Signed domains use the sign of a real scalar field: positive values are inside, negative values are outside, and zero is the boundary.

A list of conditions represents a combined feasible region. Every listed condition or signed expression must be satisfied at a sampled point for that point to be inside.

## Examples and Applications

### Example: compare a signed domain and a parametric boundary

A signed region can be paired with an explicit parametric curve when the boundary itself is important.

````python
theta = symbols("theta")

fig = figure("domain-parametric-boundary")
fig.show()
fig.domain_plot(1 - x**2 - y**2, (x, -1.5, 1.5), (y, -1.5, 1.5), boundary=False)
fig.parametric_plot((cos(theta), sin(theta)), (theta, 0, 2 * pi), style={"color": "black", "width": 3})
````

## Pitfalls

**Use symbolic Booleans, not Python Booleans.** `x > 0` is a symbolic condition; plain `True` and `False` are not meaningful plotted domains.

**Use a real scalar signed expression.** Vector, matrix, or complex-valued expressions cannot define a filled signed domain.

**Choose the right boundary form.** `boundary=False` hides the domain boundary; use [parametric_plot](parametric_plot.ipynb) for an explicit parametrized boundary curve.
