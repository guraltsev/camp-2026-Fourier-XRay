# activate

Turn on the notebook-friendly math workspace by populating the current namespace. The package profile in `math_toolkit.__init__` selects the structural notation that makes `x[i]` and `F[t](x)` work.

`math_toolkit.notebook.activate()`  
`math_toolkit.notebook.activate(namespace=scope)`


```python
x[i + 1] @Eq@ (A[i] * x[i] + b[i])
```

## Core behavior

`activate()` is the notebook entry point. It writes the managed names into the target namespace using the extension behavior selected by the root package profile. In a normal notebook session, calling it once near the top gives you SymPy, convenience symbols, function heads, `Help`, and the structural notation shown throughout the docs.


## Common patterns

Most notebooks should call `activate()` once near the top and then work directly with the injected names. Re-run it when you intentionally want to refresh the managed names.

## Options and Details

Pass `namespace=` when activation should write into an explicit mapping instead of the current notebook globals. This is useful for helper code or demonstrations that should not change the working notebook namespace.

Enabled notation and SymPy patches are selected by imports and `patch()` calls in `src/math_toolkit/__init__.py`. Optional modules may extend core objects when that profile imports them.

```python
scope = {"custom": 123}
math_toolkit.notebook.activate(namespace=scope)
scope["x"][scope["i"] + 1] @Eq@ (scope["x"][scope["i"]] + 1)
```

## Semantics

`activate()` overwrites the managed notebook bindings such as `x`, `F`, and the public math names in the target namespace, but it leaves unrelated names alone.

## Pitfalls

`activate()` does not clear the whole workspace. It refreshes managed names, while unrelated variables and other session state remain in place.
