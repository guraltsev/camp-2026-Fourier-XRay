# temperature_plot

`temperature_plot(...)` samples a real scalar field $z=f(x,y)$ as a heatmap in a live figure.

Preferred form:

```python
fig = figure("field")
fig.show()
fig.temperature_plot(expr, (x, xmin, xmax), (y, ymin, ymax))
```

````python
x, y, a = symbols("x y a")

fig = figure("temperature-quick-payoff")
fig.show()
fig.parameters({a: 1.0})
fig.temperature_plot(
    sin(a * x) * cos(y),
    (x, -pi, pi),
    (y, -pi, pi),
    label="sin(a x) cos(y)",
)
````

## Core behavior

`fig.temperature_plot(expr, x_domain, y_domain)` plots one scalar value at each sampled point of a two-dimensional domain. Use domain symbols such as `x, y` for the default visible window, or finite tuples such as `(x, -2, 2)` and `(y, -1, 3)` to set sampling bounds.

The command returns a `PlotHandle`, so heatmaps can be labeled, styled, parameter-updated, shown through their parent figure, or removed like other plots.

## Common patterns

**Explicit rectangular domains.** Finite domain tuples are the clearest way to state the sampled window.

**Parameters.** Symbols other than `x` and `y` can become sliders through `fig.parameters(...)`.

**Named updates.** Reusing `name=` updates an existing heatmap in the figure.

````python
fig = figure("temperature-parameters")
fig.show()
fig.parameters({a: {"value": 0.5, "min": 0.1, "max": 2.0, "step": 0.1}})

fig.temperature_plot(
    exp(-a * (x**2 + y**2)) * cos(3 * x),
    (x, -2, 2),
    (y, -2, 2),
    name="gaussian-field",
)
````

## Options and Details

`samples=` accepts one integer for both axes or `(x_samples, y_samples)` for separate axis densities.

`style=` supports heatmap-oriented keys such as `colorscale`, `opacity`, `visible`, `showscale`, `zmin`, `zmax`, and `zsmooth`.

`label=` controls the displayed legend/control label. `name=` controls identity for retrieval and in-place updates.

````python
fig = figure("temperature-style")
fig.show()
fig.temperature_plot(
    x**2 - y**2,
    (x, -2, 2),
    (y, -2, 2),
    samples=(80, 60),
    style={"colorscale": "Viridis", "opacity": 0.85, "showscale": True},
)
````

## Semantics

A heatmap expression must evaluate to one real scalar value at each sampled `(x, y)` point. Matrix-valued, vector-valued, complex-valued, or unevaluated symbolic results are not scalar heatmap data.

Changing `samples=` changes the grid resolution, not the mathematical expression. Changing `zmin` and `zmax` changes color scaling, not sampled values.

## Examples and Applications

### Example: combine heat and contour views

A heatmap can share a figure with contours that emphasize level-set geometry.

````python
fig = figure("temperature-with-contour")
fig.show()
field = x**2 + y**2
fig.temperature_plot(field, (x, -2, 2), (y, -2, 2), name="field")
fig.contour_plot(field, (x, -2, 2), (y, -2, 2), name="levels", style={"contour_color": "white"})
````

## Pitfalls

**Use two domain variables.** A heatmap needs both x and y domains unless you are updating an existing named heatmap.

**Use real scalar fields.** For regions use [domain_plot](domain_plot.ipynb); for two-coordinate curves use [parametric_plot](parametric_plot.ipynb).

**Remember that color is scale-dependent.** `zmin` and `zmax` affect visual interpretation, so choose them deliberately when comparing heatmaps.
