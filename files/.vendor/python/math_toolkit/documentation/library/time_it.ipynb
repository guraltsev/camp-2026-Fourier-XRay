# time_it

Measure how long a notebook calculation takes without wrapping it in setup code.

`with time_it(): ...`  
`with time_it("label"): ...`


```python
with time_it("square table"):
    squares = [n**2 for n in range(100)]
```

## Core behavior

`time_it(...)` is a context manager for quick wall-clock measurements. It records the start time before the block runs, then prints the label and elapsed seconds when the block exits.

## Common patterns

Use the default label when you only need a quick local measurement.

```python
with time_it():
    total = sum(range(1000))
```

## Options and Details

Pass a label that names the calculation being measured. The output always uses seconds with six digits after the decimal point, so repeated runs are easy to compare visually.

```python
expr = (x + 1)**4

with time_it("symbolic substitution"):
    value = expr.subs(x, 2)
```

## Semantics

`time_it` uses Python's high-resolution performance counter. It prints when the block exits, including when the block raises an exception, which makes it useful around exploratory code that may still fail.

## Pitfalls

`time_it` is a lightweight stopwatch, not a statistical benchmark. For noisy, very short, or performance-critical comparisons, repeat the measurement or use Python's dedicated benchmarking tools.
