# parametric_plot

`parametric_plot(...)` samples two coordinate expressions as a plane curve $(x(t), y(t))$.

Preferred form:

```python
fig = figure("curve")
fig.show()
fig.parametric_plot((x_expr, y_expr), (t, tmin, tmax))
```

````python
t, a = symbols("t a")

fig = figure("parametric-quick-payoff")
fig.show()
fig.parameters({a: 0.6})
fig.parametric_plot((cos(t), a * sin(t)), (t, 0, 2 * pi), label="ellipse")
````

## Core behavior

`fig.parametric_plot((x_expr, y_expr), (t, tmin, tmax))` draws a two-dimensional curve by sampling both coordinate expressions over the same parameter interval.

The coordinate pair may be a tuple, list, or SymPy `Matrix` with exactly two entries. The returned `PlotHandle` can update label, style, parameters, display, or removal.

## Common patterns

**Closed curves.** Use intervals such as `(t, 0, 2*pi)` for circles, ellipses, and periodic curves.

**Named parameters.** Symbols other than the curve parameter can become sliders through `fig.parameters(...)`.

**Overlay with domains.** Parametric curves are useful as explicit boundaries on top of [domain_plot](domain_plot.ipynb).

````python
fig = figure("parametric-slider")
fig.show()
fig.parameters({a: {"value": 0.25, "min": 0.0, "max": 0.6, "step": 0.05}})
fig.parametric_plot(
    ((1 + a * cos(5 * t)) * cos(t), (1 + a * cos(5 * t)) * sin(t)),
    (t, 0, 2 * pi),
    samples=800,
)
````

## Options and Details

`samples=` controls the number of sampled parameter values.

`style=` supports line keys such as `color`, `width`, `opacity`, `visible`, and `dash`.

`name=` identifies the plot inside the figure for retrieval and in-place updates. `label=` changes the displayed legend/control label.

````python
fig = figure("parametric-style")
fig.show()
fig.parametric_plot(
    (cos(3 * t), sin(2 * t)),
    (t, 0, 2 * pi),
    name="lissajous",
    style={"color": "purple", "width": 3, "dash": "solid"},
)
````

## Semantics

A parametric plot needs exactly two real coordinate expressions and one explicit parameter interval. The parameter is not an x-axis domain variable; it is the sampling variable that produces both coordinates.

The curve can cross itself or run backward as the parameter changes. The plotted geometry follows the sampled coordinate pair, not a single-valued function $y=f(x)$.

## Examples and Applications

### Example: phase portrait curve

When two quantities are both functions of time, parametric plotting shows their relation directly.

````python
fig = figure("parametric-phase")
fig.show()
fig.parametric_plot((cos(t), -sin(t)), (t, 0, 2 * pi), label="harmonic oscillator phase")
````

## Pitfalls

**Give an explicit parameter interval.** `fig.parametric_plot((cos(t), sin(t)))` is incomplete; write `(t, 0, 2*pi)`.

**Use exactly two coordinates.** One coordinate belongs in [plot](plot.ipynb); scalar fields belong in [temperature_plot](temperature_plot.ipynb) or [contour_plot](contour_plot.ipynb).

**Do not confuse parameter and x-domain.** The first coordinate expression supplies plotted x-values; the parameter interval supplies sample values.
