# contour_plot

`contour_plot(...)` samples a real scalar field $z=f(x,y)$ and draws its level curves.

Preferred form:

```python
fig = figure("levels")
fig.show()
fig.contour_plot(expr, (x, xmin, xmax), (y, ymin, ymax))
```

````python
x, y, a = symbols("x y a")

fig = figure("contour-quick-payoff")
fig.show()
fig.parameters({a: 1.5})
fig.contour_plot(x**2 + a * y**2, (x, -2, 2), (y, -2, 2), label="elliptic levels")
````

## Core behavior

`fig.contour_plot(expr, x_domain, y_domain)` draws level curves for one scalar field over a two-dimensional domain. Use symbols for the default window or finite domain tuples for explicit sampling bounds.

The returned `PlotHandle` can update the label, style, parameters, display, or remove the contour plot.

## Common patterns

**Level-set geometry.** Contours are a good first view for equations such as `x**2 + y**2` or potential-energy surfaces.

**Overlay on heatmaps.** Put `temperature_plot(...)` and `contour_plot(...)` in the same figure when color and level geometry both matter.

**Named updates.** Reuse `name=` to update the existing contour plot in place.

````python
fig = figure("contour-overlay")
fig.show()
field = sin(x) * cos(y)
fig.temperature_plot(field, (x, -pi, pi), (y, -pi, pi), name="field", style={"opacity": 0.75})
fig.contour_plot(field, name="levels", style={"contour_color": "black", "contour_width": 2})
````

## Options and Details

`samples=` accepts one integer for both axes or `(x_samples, y_samples)`.

`style=` supports scalar-field color keys such as `colorscale`, `opacity`, `visible`, `showscale`, `zmin`, and `zmax`, plus contour-specific keys such as `contour_color`, `contour_width`, and `line_smoothing`.

`fig.parameters(...)` creates sliders for symbols that are not the x or y domain variables.

## Semantics

A contour expression must evaluate to one real scalar value at each sampled `(x, y)` point. The contour lines visualize where that sampled scalar field crosses common levels.

Contours approximate level sets numerically. Higher `samples=` can reveal finer structure, but it does not change the symbolic field.

## Examples and Applications

### Example: a saddle field

Contours show the sign and symmetry of a saddle without requiring a three-dimensional surface.

````python
fig = figure("contour-saddle")
fig.show()
fig.contour_plot(x**2 - y**2, (x, -2, 2), (y, -2, 2), samples=100, style={"contour_width": 2})
````

## Pitfalls

**Contours are not filled regions.** Use [domain_plot](domain_plot.ipynb) when the goal is to fill where a condition is true.

**Use scalar fields.** Use [parametric_plot](parametric_plot.ipynb) for two-coordinate curves.

**Give enough samples.** Sparse grids can miss tight level-set structure.
