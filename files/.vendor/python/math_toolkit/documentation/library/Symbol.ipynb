# Symbol

Use atomic names like `x`, `theta`, or `speed` as the leaves of larger symbolic expressions.

`x`  
`theta`  
`Symbol("speed")`


```python
speed = Symbol("speed")
speed @Eq@ (x + theta)
```

## Core behavior

A symbol is an atomic quantity. It is the thing you add, differentiate with respect to, place into indices, or feed into function calls. In `math_toolkit`, atomic names stay plain: when you want structure, you write that structure with notation such as `x[i]` or `F(x)`.


## Common patterns

Atomic names are useful on their own, but the real payoff is that they stay clean when you build larger objects.


```python
diff(F(x) + speed * x, x)
```

```python
x[i + 1] @Eq@ (x[i] + speed)
```

## Options and Details

A symbol exposes intent getters for code that wants to be explicit about the role of a name. `x.symbol` returns `x` unchanged when the atomic-symbol intent matters. `x.indexedbase` returns an indexed base named `x`, so `x.indexedbase[i]` declares that the same visible name is being used as an indexed base.

## Semantics

Use a symbol when the quantity itself is atomic. Use `Indexed` notation when the quantity carries labels or components, and use `Function` notation when the object is a callable map. After activation, multi-letter names such as `speed` and Greek names such as `theta` render naturally in notebook output, but they are still plain atomic symbols.

## Pitfalls

`x_1` is rejected on purpose. It hides indexing inside a name string, which makes later symbolic work harder. Write `x[1]` or `x[i]` instead.

`x[i]` is no longer a plain symbol; it moves to the `Indexed` topic because the brackets are real structure.


```python
try:
    Symbol("x_1")
except ValueError as err:
    print(err)
```
