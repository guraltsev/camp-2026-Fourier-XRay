# Indexed

Use bracket notation to keep bases and indices separate, so recurrences, components, and indexed function heads stay readable.

`x[i]`  
`x[i, j]`  
`F[t]`


```python
x[t + 1] @Eq@ (A[t] * x[t] + b[t])
```

## Core behavior

Indexed notation adds structure without encoding it into a name string. `x[t + 1]` remembers both the base `x` and the symbolic index `t + 1`. That makes the notation useful for recurrences, components, and labeled families.


## Common patterns

The same bracket syntax covers scalar-like arrays, multi-index objects, and function heads that vary with a label.


```python
x[i, j] + y[i, j]
```

```python
F[t]
```

```python
x[t + 1] @Eq@ F[t](x[t])
```

## Options and Details

When you want the base itself before indexing, use the symbol's explicit indexed-base view. `x.indexedbase` is an indexed base named `x`, and `x.indexedbase[i]` carries the same visible indexed notation as `x[i]` while making the indexed-base intent explicit.

## Semantics

`x[i]` and `F[t]` both belong here because both are bracket-structured objects. The important difference is that `F[t]` is still a function head. It becomes an `Expression` only after you apply it, as in `F[t](x)`. Repeated brackets extend the structural index list, so `F[t][i]` and `F[t, i]` describe the same indexed head.

## Pitfalls

Do not smuggle several indices into one string. The toolkit expects each index to stay structural and separate.


```python
try:
    x["i,j"]
except ValueError as err:
    print(err)
```
