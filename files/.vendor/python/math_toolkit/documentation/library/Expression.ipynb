# Expression

Use expressions for symbolic work you want to manipulate: residuals, substitutions, differentiation, and equations built from structured parts.

`x + y`  
`x[t + 1] - A[t] * x[t] - b[t]`  
`F[t](x)`


```python
x[t + 1] - A[t] * x[t] - b[t]
```

## Core behavior

An expression is something you can manipulate algebraically. It can be added, substituted into, differentiated, simplified, or placed inside equations. Applied indexed function calls such as `F[t](x)` live here because they behave like ordinary symbolic expressions once the call has happened.


## Common patterns

Expressions are where the notation starts doing real work. You can carry labels through substitutions, differentiate with respect to arguments, and combine structured pieces into larger formulas.


```python
(x[t + 1] - F[t](x[t]) - b[t]).subs(t, n)
```

```python
diff(F[t](x), x)
```

```python
F[t + 1](x) @Eq@ G[t](F[t](x))
```

## Semantics

The applied form `F[t](x)` belongs here, but the uncalled head `F[t]` belongs under `Indexed`. Atomic names such as `x` and `theta` can appear inside expressions, yet they still have their own `Symbol` page because they are the simplest starting point. Reach for this page when you want the algebraic behavior of a composed object.


## Pitfalls

Do not confuse a function head with its applied value. `F[t]` can be indexed again or called, but it is not yet the same kind of object as `F[t](x)`.
