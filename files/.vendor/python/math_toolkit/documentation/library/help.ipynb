# help

Check an object's runtime kind and open the notebook that explains it.

`Help(obj)`  
`Help(obj, raw=True)`  
`math_toolkit.util.jupyter_document_link(path, text=...)`


```python
Help(F[t](x))
```

## Core behavior

`Help(...)` shows two things together: what the object really is at runtime, and which notebook topic explains how to work with it. That is especially useful when the runtime type is narrower than the documentation page, as with indexed function heads and applied indexed calls.


## Common patterns

Call `Help(obj)` in a notebook when you want the rendered runtime kind and clickable documentation link next to the expression you are studying.

## Options and Details

Use `raw=True` when you want plain text output for logs, tests, or copy-and-paste notes. Use `math_toolkit.util.jupyter_document_link(...)` when notebook code needs to build a link to a known documentation path directly.

```python
Help(F[t], raw=True)
```

```python
Help(object(), raw=True)
```

```python
from math_toolkit.util import jupyter_document_link

jupyter_document_link(
    "/_math_toolkit/documentation/library/Symbol.ipynb",
    text="Open the Symbol page manually",
)
```

## Semantics

`Help` reports the documentation attached to the object you pass it. If no documentation is attached, it says so directly. It is a runtime truth tool, not a search box.


## Pitfalls

Pass the actual object you want to understand. `Help` will not guess a topic from a string name or from a loose description.
