# Num

`Num` converts symbolic expressions into numeric objects. It can provide numerical values of constants, numeric implementation of functions, and numeric solutions to problems or equations.

`Expression >> Num`  
`Expression >> Num(var=..., options...)`  


```python
sin(2) + sqrt(2**2+5) >> Num
```

```python
f = (sin(a*b+c) >> Num(var=(a, b, c)))
f( 2*pi >> Num, 5, pi/6 >> Num )
#TODO: By default make >>Num safe: all inputs are wrapped into Num themselves to avoid the issue of having to numerify symbolic objects)
```

## Core behavior

Build the symbolic expression first, then cross the boundary with `Num`.


Simple symbolic expressions.

```python
sin(2)+sin(3)+sin(4) >> Num
```

More symbolic expressions

```python
_1 = Sum(k**2,(k,0,10))
_2 = _1 >> Num
md(r"$${{_1}}={{_2}}$$")
```

```python
_1 = Integral(E**(-x**2), (x , -oo, +oo)) 
_2 = _1 >> Num
md(r"""
$${{_1}}={{_2}}$$
$${{sqrt(pi)}}={{sqrt(pi)>>Num}}$$
""")
```

If the expression has free variables, its numeric form is a function. Arguments are ordered alphabetically, by default.

```python
f = sin(x-2*y) >> Num 
f(1,2)
```

## Argument Roles

Free symbols become scalar arguments by default. Explicit `Num(var=...)` arguments set the call order. Otherwise arguments are ordered alphabetically.


```python
f = x**y >> Num
_1 = f(2,3)
g = x**y >> Num(var=(y, x))
_2 =  g(2,3)
md(r"""$$ f(2,3) = {{_1}} \qquad g(2,3) = {{_2}}$$""") 
```

`Num(var=...)` knows array access

```python
5
_+1
```

```python
f = (x[i] + x[i + 1]) >> Num(var={x: Num.Array(1) , i: Num.Index})
f([0,1,2,3,4],2)
```

```python
ordered_f = (x**2 + y**2) >> Num(var=(y, x))
md("""Explicit order gives {{ ordered_f }} and ordered_f(3, 4) = {{ ordered_f(3, 4) }}.""")
```

## Symbolic Preparation

`Num` does not substitute values and does not automatically expose held definitions. Use `subs` before `Num`, and use `Unhold` or `UnholdAll` when you want held definitions to become ordinary symbolic bodies first.


```python
prepared_value = sin(x) >> subs((x, 2)) >> Num
prepared_function = Hold(x + 1) >> UnholdAll >> Num

md("""Substitution before Num gives {{ prepared_value }}.
Unholding before Num gives prepared_function(2) = {{ prepared_function(2) }}.""")
```

## Shape Policy

Numeric output shape is `sample shape + mathematical output shape`. A vector-valued expression evaluated on a grid therefore has one leading sample axis and one final vector axis.


```python
import numpy as np

grid = np.array([1, 2, 3])
F = Matrix([x, x**2]) >> Num(var=x)
values = F(grid)

md("""The vector output shape is {{ values.shape }} and the second sample is {{ values[1] }}.""")
```
