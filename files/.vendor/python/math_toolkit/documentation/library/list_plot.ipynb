# list_plot

`list_plot(...)` draws a finite set of points from values, explicit coordinates, or integer-indexed expressions.

Preferred forms:

```python
fig = figure("discrete")
fig.show()
fig.list_plot([2, 3, 5, 7])
```

```python
fig.list_plot(n**2, (n, 0, 8))
```

````python
n = symbols("n", integer=True)

fig = figure("list-plot-quick-payoff")
fig.show()
fig.list_plot([2, 3, 5, 7, 11], label="prime samples")
````

## Core behavior

`fig.list_plot(source)` adds or updates one discrete point set. The source can be a finite sequence of y-values, explicit `(x, y)` pairs, a scalar expression sampled at integer indices, a two-coordinate expression, or a pair of coordinate expressions.

Value lists own their index domain. `fig.list_plot([2, 3, 5])` plots points `(0, 2)`, `(1, 3)`, and `(2, 5)`. Explicit coordinate pairs keep their x-values instead of replacing them with positions in the list.

````python
fig = figure("list-plot-values-and-points")
fig.show()

fig.list_plot([2, 3, 5, 7], name="indexed", label="indexed values")
fig.list_plot([(0, 1), (2, 3), (5, 2)], name="explicit", label="explicit xy")
````

## Common patterns

**Indexed sequences.** Use `fig.list_plot(values)` when the x-coordinate is the position of each value.

**Point clouds.** Use `fig.list_plot([(x0, y0), (x1, y1), ...])` when the data already carries x-coordinates.

**Discrete expressions.** Use `fig.list_plot(expr, (n, start, stop))` for a half-open integer range. The expression is evaluated at `start, start + 1, ..., stop - 1`.

**Expression pairs.** Use `fig.list_plot([x_expr, y_expr], (n, start, stop))` for a point cloud whose coordinates are both discrete expressions.

````python
fig = figure("list-plot-expression-pair")
fig.show()

fig.list_plot([cos(n), sin(n)], (n, 0, 32), label="discrete unit circle")
````

## Options and Details

`name=` identifies the point set inside the figure for retrieval and in-place updates. `label=` changes the displayed legend/control label.

`fig.parameters(...)` creates sliders for symbols that are not the integer index. The compact form is `{symbol: value}`; dictionary specs can also supply `min`, `max`, `step`, and `label`.

`style=` accepts the shared line-style keys `color`, `width`, `opacity`, `visible`, and `dash`. List plots render as markers by default so the finite sample set remains visible as discrete points.

````python
scale = symbols("scale")

fig = figure("list-plot-parameter-style")
fig.show()
fig.parameters({scale: {"value": 0.5, "min": 0.0, "max": 2.0, "step": 0.1}})
fig.list_plot(
    scale * n**2,
    (n, 0, 8),
    style={"color": "crimson", "width": 2},
    label="scaled squares",
)
````

## Semantics

A list plot is not a continuous curve. It samples or receives a finite set of points and sends those points to the renderer as a scatter trace.

Expression sources use integer-index sampling. With an explicit index spec such as `(n, 0, 10, 2)`, the sampled indices are `0, 2, 4, 6, 8`. The upper endpoint is excluded, matching Python's `range(...)` convention.

With a bare index symbol such as `fig.list_plot(n**2, n)`, the current x-axis view supplies the visible integer indices. Panning or zooming the x-axis resamples the integer values in view. Value inputs and explicit index specs keep their fixed sample domain when the view changes.

````python
fig = figure("list-plot-view-aware")
fig.show()
fig.list_plot(n**2, n, name="visible-squares", label="visible integer squares")
fig.view.range = {"x": (-3.2, 4.2), "y": (-1, 18)}
````

## Examples and Applications

### Example: compare discrete and continuous behavior

A list plot can sit beside an ordinary curve when the distinction between sampled integer values and a continuous model matters.

````python
fig = figure("list-plot-discrete-continuous")
fig.show()

fig.plot(x**2, (x, 0, 8), label="continuous x^2", style={"color": "gray", "dash": "dash"})
fig.list_plot(n**2, (n, 0, 9), label="integer samples", style={"color": "royalblue"})
````

## Pitfalls

**Do not add an index argument to value inputs.** `fig.list_plot([2, 3, 5], n)` is invalid because value lists already use their own integer positions.

**Use an index symbol or full index spec for expressions.** `fig.list_plot(n**2)` is incomplete; write `fig.list_plot(n**2, n)` or `fig.list_plot(n**2, (n, 0, 10))`.

**Use exactly two coordinates for point expressions.** Scalar outputs become `(n, y)` points. Two-coordinate outputs become `(x, y)` points. Higher-dimensional outputs are not list plots in this phase.

**Choose the plot kind by mathematical shape.** Use [plot](plot.ipynb) for continuous scalar curves and [parametric_plot](parametric_plot.ipynb) for continuous coordinate curves.
