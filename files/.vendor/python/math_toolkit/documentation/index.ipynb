# math_toolkit documentation

`math_toolkit` is a notebook toolkit for writing mathematics in Python without giving up readable notation. It helps with symbolic expressions, named functions, indexed notation, equation building, readable output, plotting, and the step from symbolic work to numerical experiments.

Start a notebook by importing and activating the toolkit:


```python
import math_toolkit
math_toolkit.activate()
```

After activation, the usual SymPy names and the toolkit's notebook helpers are available to use in the notebook. Use the tutorials below when you want a reusable pattern, use the worksheet when you want to see a longer mathematical investigation, and use `Help(obj)` when you want the reference page for a specific object or function.
