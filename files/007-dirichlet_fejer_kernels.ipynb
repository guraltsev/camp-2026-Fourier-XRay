# Dirichlet and Fejer kernels

The square-wave inversion notebook shows a strange feature of Fourier partial sums: adding more modes improves the approximation, but the oscillations near a jump do not simply disappear.

The reason is already visible in the kernels. The Dirichlet kernel is the kernel for ordinary Fourier partial sums. The Fejer kernel is the kernel for the average of the first partial sums.

This worksheet animates both kernels so we can compare how their mass moves as the number of modes grows.

## The kernels

On a one-period interval, the Dirichlet kernel is

$$
D_N(x)=\sum_{k=-N}^{N} e^{2\pi i kx}=1+2\sum_{k=1}^{N}\cos(2\pi kx).
$$

The Fejer kernel is the average of the Dirichlet kernels:

$$
F_N(x)=\frac{1}{N+1}\sum_{j=0}^{N}D_j(x)
     =1+2\sum_{k=1}^{N}\left(1-\frac{k}{N+1}\right)\cos(2\pi kx).
$$

So $D_N$ gives every frequency up to $N$ the same weight, while $F_N$ tapers the high frequencies.

```python
# Symbolic definitions for the two kernel sums.
D = Function("D")(x, N) @EqDef@ (
    1 + 2 * Sum(cos(2 * pi * k * x), (k, 1, N))
)

F = Function("F")(x, N) @EqDef@ (
    1 + 2 * Sum((1 - k / (N + 1)) * cos(2 * pi * k * x), (k, 1, N))
)

r"""
We will animate
$$
D_N(x) = {{ D(x,N) }}
$$
and
$$
F_N(x) = {{ F(x,N) }}.
$$
""" >> md
```

## Animate the kernels

Press play and watch the side lobes. Both kernels build a tall peak at the origin, but they behave very differently away from the origin.

```python
kernel_fig = figure("Dirichlet and Fejer kernels", new=True)
kernel_fig.view.home_range = {"x": (-0.5, 0.5), "y": (-25, 85)}
kernel_fig.show()

kernel_fig.parameters({
    N: {"value": 1, "min": 1, "max": 80, "step": 1},
})

kernel_fig.plot(
    D(x, N),
    x,
    name="Dirichlet",
    label=r"$D_N(x)$",
    style={"color": "#2563eb", "width": 2},
    samples=2001,
)
kernel_fig.plot(
    F(x, N),
    x,
    name="Fejer",
    label=r"$F_N(x)$",
    style={"color": "#dc2626", "width": 2},
    samples=2001,
)
```

## The coefficient weights

The two kernels differ because they put different weights on the Fourier modes. The Dirichlet kernel keeps every mode from $-N$ to $N$ with weight $1$. The Fejer kernel uses triangular weights:

$$
1-\frac{|k|}{N+1}, \qquad |k|\le N.
$$

Animate the weights and compare this picture with the kernel animation above.

```python
weight_fig = figure("Fourier-mode weights", new=True)
weight_fig.view.home_range = {"x": (-82, 82), "y": (0, 1.1)}
weight_fig.show()

weight_fig.parameters({
    N: {"value": 1, "min": 1, "max": 80, "step": 1},
})

dirichlet_weight = Piecewise((1, Abs(k) <= N), (0, True))
fejer_weight = Piecewise((1 - Abs(k) / (N + 1), Abs(k) <= N), (0, True))

weight_fig.list_plot(
    dirichlet_weight,
    (k, -80, 81),
    name="Dirichlet weights",
    label=r"Dirichlet weights",
    style={"color": "#2563eb", "width": 2, "opacity": 0.45},
)
weight_fig.list_plot(
    fejer_weight,
    (k, -80, 81),
    name="Fejer weights",
    label=r"Fejer weights",
    style={"color": "#dc2626", "width": 2, "opacity": 0.85},
)
```
